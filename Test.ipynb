```python
from shapely.geometry import Point, LineString

def perpendicular_distance(point_geom, line_geom):
    # Ensure the geometries are of the correct type
    if not isinstance(point_geom, Point) or not isinstance(line_geom, LineString):
        raise ValueError("Input geometries must be a Point and a LineString.")

    # Project the point onto the line
    projected_point = line_geom.interpolate(line_geom.project(point_geom))

    # Calculate the distance between the original point and the projected point
    distance = point_geom.distance(projected_point)

    return distance

# Example usage:
# Create a Point geometry representing a node
node_geom = Point(1, 1)

# Create a LineString geometry representing a linestring
linestring_geom = LineString([(0, 0), (2, 0)])

# Calculate the perpendicular distance
distance = perpendicular_distance(node_geom, linestring_geom)

print(f"Perpendicular distance: {distance}")
```

```python
import geopandas as gpd

def find_nodes_without_intersects(nodes_shapefile, pipes_shapefile):
    # Read shapefiles
    nodes = gpd.read_file(nodes_shapefile)
    pipes = gpd.read_file(pipes_shapefile)

    # Spatial join to find intersections
    intersection = gpd.sjoin(nodes, pipes, how='left', op='intersects')

    # Get ids of nodes that don't intersect with any pipes
    nodes_without_intersects = intersection[intersection.index_right.isna()]
    ids_without_intersects = nodes_without_intersects['id'].tolist()

    return ids_without_intersects

# Example usage:
nodes_shapefile_path = 'Data SIG/3M/6-FinallRep/Nodes.shp'
pipes_shapefile_path = 'Data SIG/3M/6-FinallRep/Pipes.shp'

result = find_nodes_without_intersects(nodes_shapefile_path, pipes_shapefile_path)
print("Node IDs without intersects:", result)
```

```python
import geopandas as gpd

def find_nodes_without_intersects(nodes_shapefile, pipes_shapefile, nodes_id_column):
    # Read shapefiles
    nodes = gpd.read_file(nodes_shapefile)
    pipes = gpd.read_file(pipes_shapefile)

    # Reset the index to ensure 'id' is a regular column
    nodes = nodes.reset_index(drop=True)

    # Spatial join to find intersections
    intersection = gpd.sjoin(nodes, pipes, how='left', op='intersects')

    # Get ids of nodes that don't intersect with any pipes
    nodes_without_intersects = intersection[intersection[f"{nodes_id_column}_left"].isna()]
    ids_without_intersects = nodes_without_intersects[nodes_id_column].tolist()

    return ids_without_intersects

# Example usage:
nodes_shapefile_path = 'Data SIG/3M/6-FinallRep/Nodes.shp'
pipes_shapefile_path = 'Data SIG/3M/6-FinallRep/Pipes.shp'

result = find_nodes_without_intersects(nodes_shapefile_path, pipes_shapefile_path, 'id')
print("Node IDs without intersects:", result)
```

```python
import pandas as pd
import json

def json_to_csv(json_path, csv_path):
    # Read JSON file into a DataFrame
    with open(json_path, 'r') as file:
        data = json.load(file)

    # Extract relevant information
    rows = []
    for feature in data:
        properties = feature.get('properties', {})
        source_node = properties.get('sourceNode')
        target_node = properties.get('targetNode')

        if source_node is not None and target_node is not None:
            rows.append({'source': source_node, 'target': target_node})

    # Create DataFrame
    df = pd.DataFrame(rows)

    # Save DataFrame to CSV file
    df.to_csv(csv_path, index=False)

# Example usage:
json_file_path = 'Data SIG/3M/Json/Pipes.json'
csv_file_path = 'Data SIG/3M/Json/Pipes.csv'

json_to_csv(json_file_path, csv_file_path)
```

```python
import geopandas as gpd
import json
from shapely.geometry import Point

def create_shapefile(json_file, output_shapefile):
    # Read the JSON file
    with open(json_file, 'r') as f:
        dataNodes = json.load(f)

    # Create a list of dictionaries for nodes
    nodes = []
    for feature in dataNodes:
        if feature['properties']['source_1'] == 'Dummy':
            nodes.append({
                'id': feature['properties']['id'],
                'components': feature['properties']['components'],
                'source_1': feature['properties']['source_1'],
                'geometry': Point(feature['geometry']['coordinates'])
            })

    # Create a GeoDataFrame for filtered nodes
    gdf_dummy = gpd.GeoDataFrame(nodes, geometry='geometry')

    # Save to shapefile
    gdf_dummy.to_file(output_shapefile, driver='ESRI Shapefile')

# Example usage
create_shapefile('Data SIG/1704/M3M/Json/Nodes.json', 'output_shapefile.shp')
```

```python
import json
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# List of JSON files
json_files = ['Data SIG/3M/Json/Pipes.json', 'Data SIG/1402/Json/Pipes.json', 'Data SIG/1704/M3M/Json/Pipes.json', 
              'Data SIG/200211_Veolia_M3M/Json/Pipes.json', 'Data SIG/200213_3M_data/Json/Pipes.json']

# Maximum diameter limit
max_diameter_limit = 1500

# Create a single plot
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_title('Distribution of Pipe Diameters')

# Iterate through each JSON file and plot the bell curve
for i, json_file in enumerate(json_files):
    # Load JSON data
    with open(json_file, 'r') as file:
        data = json.load(file)

    # Extract diameters (check for key existence)
    diameters = [pipe['properties'].get('diametre', 0) for pipe in data]

    # Remove entries with missing or zero diameters and filter by maximum limit
    diameters = [diameter for diameter in diameters if 0 < diameter <= max_diameter_limit]

    # Check if there are valid diameters
    if diameters:
        # Compute mean and standard deviation
        mean_diameter = np.mean(diameters)
        std_diameter = np.std(diameters)

        # Create x values for the bell curve
        x = np.linspace(min(diameters), max(diameters), 100)

        # Plot the bell curve with different colors
        color = plt.cm.viridis(i / len(json_files))
        ax.plot(x, norm.pdf(x, mean_diameter, std_diameter), label=f'File {i + 1}', color=color)

# Set labels and legend
ax.set_xlabel('Diameter')
ax.set_ylabel('Density')
ax.legend()

# Show the plot
plt.show()
```

```python
import json

# Load pipes and nodes from JSON files
with open('Data SIG/3M/Json/Pipes.json', 'r') as f:
    pipes = json.load(f)

with open('Data SIG/3M/Json/Nodes.json', 'r') as f:
    nodes = json.load(f)

# Create a dictionary to store the count of each node's appearances in pipes
node_count = {}

# Count the occurrences of each node as a source or target in pipes
for pipe in pipes:
    source_node_id = pipe['properties']['sourceNode']
    target_node_id = pipe['properties']['targetNode']

    # Increment count for source node
    node_count[source_node_id] = node_count.get(source_node_id, 0) + 1

    # Increment count for target node
    node_count[target_node_id] = node_count.get(target_node_id, 0) + 1

# Find nodes not connected to two arcs
unconnected_nodes = [node['properties']['id'] for node in nodes if node_count.get(node['properties']['id'], 0) < 2]

# Print or use the list of unconnected nodes
print("Nodes not connected to two arcs:", unconnected_nodes)
print(len(unconnected_nodes))
```

```python
import json
from math import sqrt

# Load nodes from JSON file
with open('Data SIG/3M/Json/Nodes.json', 'r') as f:
    nodes = json.load(f)

# Extract coordinates for each unconnected node
unconnected_nodes_coordinates = {
    node['properties']['id']: (node['geometry']['coordinates'][0], node['geometry']['coordinates'][1])
    for node in nodes if node['properties']['id'] in unconnected_nodes
}

# Function to calculate Euclidean distance between two points
def euclidean_distance(coord1, coord2):
    return sqrt((coord1[0] - coord2[0])**2 + (coord1[1] - coord2[1])**2)

# Calculate and store the minimum distances between each pair of unconnected nodes
min_distances_between_nodes = {}

for node_id1, coord1 in unconnected_nodes_coordinates.items():
    for node_id2, coord2 in unconnected_nodes_coordinates.items():
        if node_id1 != node_id2:
            distance = euclidean_distance(coord1, coord2)

            # Update minimum distance if it's not set or the new distance is smaller
            if (node_id1, node_id2) not in min_distances_between_nodes or distance < min_distances_between_nodes[(node_id1, node_id2)]:
                min_distances_between_nodes[(node_id1, node_id2)] = distance

# Print or use the resulting minimum distances
for (node_id1, node_id2), min_distance in min_distances_between_nodes.items():
    print(f"Minimum Euclidean distance between nodes {node_id1} and {node_id2}: {min_distance:.2f} units")
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.neighbors import kneighbors_graph
from sklearn.manifold import spectral_embedding
import matplotlib.pyplot as plt

# Generate synthetic data with 3 clusters
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Construct a similarity graph using k-nearest neighbors
adjacency_matrix = kneighbors_graph(X, n_neighbors=10, mode='distance')

# Convert adjacency matrix to a numpy array
adjacency_matrix = np.asarray(adjacency_matrix)

# Compute the unnormalized Laplacian
laplacian_matrix = np.diag(np.sum(adjacency_matrix, axis=1)) - adjacency_matrix

# Compute the first two eigenvectors
eigenvalues, eigenvectors = np.linalg.eigh(laplacian_matrix)
embedding = eigenvectors[:, :2]

# Cluster in the embedded space using k-means
kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(embedding)

# Plot the results
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', edgecolors='k', s=50)
plt.title('Unnormalized Spectral Clustering')
plt.show()
```
